```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import keras
from scipy import ndimage
import numpy as np
from tensorflow.keras import layers, models, callbacks, optimizers, losses
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.layers import RandomRotation, RandomZoom, RandomCrop
from tensorflow.data import AUTOTUNE
```

```python
def create_data_loader(train_dir, test_dir, valid_dir, image_size=(224, 224), batch_size=120, limit=100):
    datagen = ImageDataGenerator(
        rescale=1./255,
    )
    
    rotation_range = 20
    zoom_range = (0.8, 1.2)

    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=rotation_range,
        zoom_range=zoom_range
    )

    train_ds = train_datagen.flow_from_directory(
        train_dir,
        target_size=image_size,
        batch_size=batch_size,
        seed=1337,
        shuffle=True,
    )

    test_ds = datagen.flow_from_directory(
        test_dir,
        target_size=image_size,
        batch_size=batch_size,
    )

    valid_ds = datagen.flow_from_directory(
        valid_dir,
        target_size=image_size,
        batch_size=batch_size,
        seed=1337,
        shuffle=False,
    ) 
    
    return train_ds, test_ds, valid_ds
```

```python
train_dir = 'Data/Train'
test_dir = 'Data/Test'
valid_dir = 'Data/Valid'

train_ds, test_ds, valid_ds = create_data_loader(train_dir, test_dir, valid_dir, limit=100)

# Load Images for Visualization
images, _ = next(train_ds)

plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i])
    plt.axis('off')
plt.show()
```

```python
def create_transfer_model(input_shape, num_classes):    
    backbone = keras.applications.MobileNetV2(
        weights="imagenet",
        include_top=False,
        input_shape=input_shape,
    )
    backbone.trainable = False

    # Building the transfer model architecture
    inputs = layers.Input(input_shape)
    x = keras.applications.mobilenet_v2.preprocess_input(inputs)
    x = backbone(x)
    x = layers.Dropout(0.3)(x)
    x = layers.GlobalAveragePooling2D()(x)
    
    # Adjust the output layer to have num_classes units
    outputs = layers.Dense(num_classes, activation=None)(x)

    return keras.Model(inputs, outputs, name="transfer_model")

train_ds, test_ds, valid_ds = create_data_loader(train_dir, test_dir, valid_dir, limit=100)
```

```python
# Define model checkpoint
checkpoint_path = "model_checkpoint.keras"
checkpoint = keras.callbacks.ModelCheckpoint(checkpoint_path, 
                                             monitor='val_accuracy', 
                                             verbose=1, 
                                             save_best_only=True,
                                             mode='max')

input_shape = (224, 224, 3)
num_classes = 4

# Create and compile the model
model2 = create_transfer_model(input_shape, num_classes)

model2.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0001),
    loss=keras.losses.CategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)


# Training the model
history = model2.fit(
    train_ds,
    validation_data=valid_ds,
    epochs=30,  
    callbacks=[checkpoint]
)

# Loading the best weights
model2.load_weights(checkpoint_path)

# Evaluate the model on the test set
test_loss, test_accuracy = model2.evaluate(test_ds)
print("\nTest Accuracy:", test_accuracy)
```

```python
#Roc Curve for loss & Val Loss

plt.plot(range(len(history.history["loss"])),history.history["loss"],label="Training Loss")
plt.plot(range(len(history.history["val_loss"])),history.history["val_loss"],label="Validation Loss")
plt.legend()
```

```python
#ROC Curve for accuracy & Val Accuracy

plt.plot(range(len(history.history["accuracy"])),history.history["accuracy"],label="Training Loss")
plt.plot(range(len(history.history["val_accuracy"])),history.history["val_accuracy"],label="Validation Loss")
plt.legend()
```
